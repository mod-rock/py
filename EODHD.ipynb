## 2) EOD

```python
def priceEOD(symbol, period):
    
    mkt = general(symbol).CountryISO.Country
    url = 'https://eodhistoricaldata.com/api/eod/'+ symbol +'.' + mkt + '?api_token=' + api_key + '&period=' + period
    df = pd.read_csv(url)
    df = df.set_index(['Date'])
    df = df.apply(pd.to_numeric)
    v = df['Volume'].values
    tp = (df['Low'] + df['Close'] + df['High']).div(3).values
    
    return df.assign(vwap=(tp * v).cumsum() / v.cumsum())
    #return df
```

```python
priceEOD('AAPL', 'd')
```

## 3) Intraday

```python
def priceIntraDay(symbol, interval):
    
    mkt = general(symbol).CountryISO.Country
    url = 'https://eodhistoricaldata.com/api/intraday/'+ symbol +'.' + mkt + '?api_token=' + api_key + '&period=' + interval
    df = pd.read_csv(url)
    df['Datetime'] = pd.to_datetime(df.Datetime)
    df = df.set_index(['Datetime'])
    #df = df.apply(pd.to_numeric)

    return df
```

```python
priceIntraDay('AAPL', '5m')
```
