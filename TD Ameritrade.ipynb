```python
def price_history(ticker, period_Type, period_, freq_Type, freq_):

    full_url = 'https://api.tdameritrade.com/v1/marketdata/' + ticker.upper() + '/pricehistory?apikey=' + td_key + '&periodType=' + period_Type + '&period=' + str(period_) + '&frequencyType=' + freq_Type + '&frequency=' + str(freq_)
    
    response = requests.get(full_url)
    response_json = response.json()
    main_df = pd.DataFrame(response_json['candles'])
    df = main_df
    df['Ticker'] = ticker.upper()
    df['datetime'] = df['datetime'].apply(lambda x: datetime.fromtimestamp(x/1000).strftime("%d-%m-%Y %I:%M:%S"))
    df = df[df.columns[[6,0,1,2,3,4,5]]]
    
    return df
```

```python
def quotes(ticker):
    full_url = 'https://api.tdameritrade.com/v1/marketdata/' + ticker.upper() + '/quotes?apikey=' + td_key
    response = requests.get(full_url)
    response_json = response.json()
    df = pd.DataFrame([response_json[ticker.upper()]])
    df['quoteTimeInLong'] = df['quoteTimeInLong'].apply(lambda x: datetime.fromtimestamp(x/1000).strftime("%d-%m-%Y %I:%M:%S"))
    df['tradeTimeInLong'] = df['tradeTimeInLong'].apply(lambda x: datetime.fromtimestamp(x/1000).strftime("%d-%m-%Y %I:%M:%S"))
    df['regularMarketTradeTimeInLong'] = df['regularMarketTradeTimeInLong'].apply(lambda x: datetime.fromtimestamp(x/1000).strftime("%d-%m-%Y %I:%M:%S"))
    df['Ticker'] = ticker.upper()
    
    return df
```

```python
def price_momentum(data, window_t):
    R_t = pd.DataFrame((data.Close - data.Close.shift(window_t))/data.Close.shift(window_t))
    T = int(R_t.count())
    R_mean = float(R_t.sum()) / T
    sigma_sq = float((R_t.sub(R_mean)**2).sum()) / (T - 1)
    R_riskAdj = R_mean / (sigma_sq ** 0.5)
    #S = -window_Skip
    #R_cum = 
    df = pd.DataFrame([R_mean, R_riskAdj], index = ['ReturnMean', 'riskAdjReturn'])
    return df.T
    
```

```python
def earnings_momentum(ticker):
    ticker = ticker.upper()
    Q_earnings = yf.Ticker(ticker).quarterly_earnings
    earningsMean = Q_earnings.Earnings.mean()
    sigma_sq = float((Q_earnings['Earnings'].sub(earningsMean)**2).sum()) / 3
    sigma = sigma_sq ** 0.5
    SUE = (Q_earnings.Earnings[1] - Q_earnings.Earnings[-1]) / int(sigma)
    return SUE
```
